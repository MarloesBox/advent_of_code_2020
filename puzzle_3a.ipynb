# Advent of Code 2020 - Puzzle 3A

## Import packages

```python
import pandas as pd
```

## Constants

```python
start_hor_coordinate = 0
start_vert_coordinate = 0
fly_right = 3
fly_down = 1
```

## Load data

```python
data = pd.read_csv("data/input_puzzle_3.txt", header = None)
data.columns = ['coordinates'] # Add column name
data.head()
```

## Functions

```python
def split(word): 
    return [char for char in word]
```

```python
def hitAlmostTree(coordinates_map, x, y):
    if (coordinates_map.iloc[x, y] == "#"):
        return True
    else:
        return False
```

## Make coordinates map

```python
# Calculate the number of coordinates:
total_hor_coordinates = len(data['coordinates'].iloc[0])
total_vert_coordinates = len(data)
```

```python
# New data frame with split value columns 
coordinates_map = data["coordinates"].str.split("", n = total_hor_coordinates, expand = True)
coordinates_map.drop(0, axis = 1, inplace = True)
coordinates_map.columns = [list(range(0, total_hor_coordinates))]
coordinates_map
```

## Main

```python
total_encounterd_trees = 0
route_map = coordinates_map.copy()

# Current start position:
current_vert_coordinate = start_vert_coordinate
current_hor_coordinate = start_hor_coordinate

# Continu untill South Pole is reached:
while current_vert_coordinate < total_vert_coordinates:
    
    if hitAlmostTree(coordinates_map, current_vert_coordinate, current_hor_coordinate) == True:
        total_encounterd_trees += 1
        route_map.iloc[current_vert_coordinate, current_hor_coordinate] = 'X' # tree
    else:
        route_map.iloc[current_vert_coordinate, current_hor_coordinate] = 'O' # open field
    
    # Slope:
    current_vert_coordinate += fly_down
    current_hor_coordinate += fly_right

    # Return to the most left position when right end of the world is reached:
    if current_hor_coordinate >= total_hor_coordinates:
        current_hor_coordinate = current_hor_coordinate - total_hor_coordinates    

print("Total number of encounted trees: ", total_encounterd_trees)
```
